##Data Preparation

```python
%cd "/content/drive/MyDrive/Assignment2/"

doc = "Sequential_Dataset.txt"
text = open(doc, 'r', encoding='utf-8').read()
text = text.lower()
```

```python
text
```

```python
# create mapping of unique chars to integers
chars = sorted(list(set(text)))
char_to_int = dict((c, i) for i, c in enumerate (chars))
print(char_to_int)
```

```python
n_chars = len(text)
n_vocab= len(chars)
print("Total Characters: ", n_chars)
print("Total Vocab (Unique characters): ", n_vocab)
```

```python
#prepare the dataset of input to output pairs encoded as integers
seq_length = 20
dataX = []
dataY = []
for i in range(0, n_chars - seq_length, 1):
  seq_in = text[i:i + seq_length]
  seq_out = text[i + seq_length]
  dataX.append([char_to_int[char] for char in seq_in])
  dataY.append(char_to_int[seq_out])
n_patterns = len(dataY)
print("Total Patterns: ", n_patterns)
```

```python
dataX
```

```python
dataY
```

```python
import numpy
# reshape X to be [samples, time steps, features]
X = numpy.reshape(dataX, (n_patterns, seq_length, 1))

# normalize - rescaling the integer values
X = X / float(n_vocab)
print(X)
```

##Model Design

```python
import tensorflow as tf
from tensorflow import keras
from keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# one hot encode the output variable
y= to_categorical(dataY)
y
```

```python
# define the LSTM model
model = Sequential()
model.add(LSTM(7, input_shape=(X.shape[1], X.shape[2])))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation="softmax"))
model.compile(loss= "categorical_crossentropy", optimizer="adam")
# define the checkpoint
filepath ="weights-improvement-{epoch:02d}-{loss:.4F}.hdfs"
checkpoint = ModelCheckpoint(filepath, monitor ="loss", verbose=1, save_best_only=True, mode="min")
callbacks_list = [checkpoint]
```

##Training

```python
epochs = 15
batch_size=  150
model.fit(X, y, epochs=epochs, batch_size=batch_size, callbacks=callbacks_list)
```

##Evaluation

```python
#load the network weights
file_weight="weights-improvement-06-3.4506.hdfs"
model.load_weights (file_weight)
model.compile(loss="categorical_crossentropy", optimizer="adam")

int_to_char = {i:char for i, char in enumerate(chars)}

# generate a random seed
print("Size of DataX:", len(dataX))
start = numpy.random.randint(0, len(dataX)-1)
print("Start from:", start)
pattern = dataX[start]
print("Seed:")
sentence = [int_to_char[value] for value in pattern]
print("\"", ''.join(sentence), "\"")
```

```python
length = 5
final = []
for i in range(length):
    x = numpy.reshape(pattern, (1, len(pattern), 1))
    x = x / float(n_vocab)
    prediction = model.predict(x, verbose=0)

    # Get the predicted values with the highest probabilities
    indices = numpy.argsort(prediction[0])[-1:]

    # Convert the indices to characters and join
    results = [int_to_char[index] for index in indices]
    final.append(''.join(results))

    # Update the pattern with both predicted indices
    pattern.extend(indices)

    # Remove the first characters from the seed sequence
    pattern = pattern[1:len(pattern)]
print(final)
```

```python
for i in final:
  sentence.append(i)
print(sentence)
```
